```python
#B
n = int(input())
arr = list(map(int,input().split()))
heap = []

for i in range(len(arr)):
    a = arr[i]
    arr[i] = -1
    if heap == []:
        heap.append((a,i))
        continue
    while(heap and heap[-1][0] > a):
        a_t, i_t = heap.pop()
        arr[i_t] = i
    
    heap.append((a,i))
    
print(*arr)
```

```python
#C
n, k = map(int,input().split())
arr = list(map(int,input().split()))

deck = ["#"]*k
start = 0
end = -1

for i in range(k):
    a = arr[i]  
    while (deck[end]!="#" and a < deck[end]):
        deck[end] = "#"
        end = (end-1)%k
    
    end = (end+1)%k
    
    deck[end] = a
    
print(deck[start])

for i in range(k,n):
    if arr[i-k] == deck[start]:
        deck[start] = "#"
        start = (start+1)%k
        
    a = arr[i]  
    while (deck[end] != "#" and a < deck[end]):
        deck[end] = "#"
        end = (end-1)%k
    
    end = (end+1)%k
    deck[end] = a
    
    print(deck[start])
```

```python
#D
arr = list(input().split())

heap = []

for a in arr:
    if a == "+":
        heap.append(heap.pop() + heap.pop())
    elif a == "-":
        heap.append(-heap.pop() + heap.pop())
    elif a == "*":
        heap.append(heap.pop() * heap.pop())
    else:
        heap.append(int(a))
        
print(heap[-1])
```

```python
#F

n = int(input())

heap = []
priority = {}
sk_op = 0

for i,j in enumerate(input()):
    priority[j] = i
    if sk_op == 0 and  j in "([{":
        sk_op = j

op = 0
inverse = {"(":")", "[":"]", "{":"}"}

ans = input()
print(ans, end = "")

for i in ans:
    if i in "([{":
        op += 1
        heap.append(i)
    else:
        heap.pop()

        
m = n//2
for i in range(n-len(ans)):
    if op < m:
        if heap and priority[inverse[heap[-1]]] < priority[sk_op]:
            print(inverse[heap[-1]], end ="")
            heap.pop()
        else:
            print(sk_op, end = "")
            op += 1
            heap.append(sk_op)
    else:
        print(inverse[heap[-1]], end ="")
        heap.pop()
```

```python
#J

n, b = map(int,input().split())
arr = [0]*n

for i, h in enumerate(map(int,input().split())):
    arr[i] = [h,0]
    
for i,w in enumerate(map(int,input().split())):
    arr[i][1] = w
    
arr.sort()

if n == 1 or arr[0][1] >= b:
    print(0)
else:
    l = 0
    r = 1
    height = arr[-1][0] - arr[0][0]
    deq = []
    width = arr[0][1]
    dd = 0


    for r in range(1,n):
        width += arr[r][1]
        if arr[r][1] >= b:
            print(0)
            dd = -1
            break
        while len(deq)>dd and deq[-1] < arr[r][0] - arr[r-1][0]:
            deq.pop()
        deq.append(arr[r][0] - arr[r-1][0])


        while(width >= b):
            height = min(height, deq[dd])

            if len(deq)>dd and deq[dd] == arr[l+1][0] - arr[l][0]:
                dd += 1

            width -= arr[l][1]
            l += 1


    if dd != -1:       
        print(height)
    
```

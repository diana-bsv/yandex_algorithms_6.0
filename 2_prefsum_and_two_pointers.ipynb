```python
#C

import tracemalloc
tracemalloc.start()

f = open("37.txt", 'r')
n, rad = map(int, f.readline().split())

arr = list(map(int, f.readline().split()))
f.close()
l = 0
r = 1

ans = 0

for l in range(n-1):
    while (r < n)and(arr[r]-arr[l] <= rad):
        r += 1
    if r == n-1 and (arr[r]-arr[l] <= rad):
        break
        
    ans += n - r
    
print(ans)


print(tracemalloc.get_traced_memory()[1]*1e-6)

# stopping the library
tracemalloc.stop()
```

```python
#F

n = int(input())
arr = list(map(int, input().split()))

dump = 0
slump = 0
ans = 0

for a in range(len(arr)-2):
    dump += arr[a]
    slump += arr[a+1]*dump
    ans += arr[a+2]*slump
    
    if ans >= 1000000007:
        ans = ans%1000000007
print(ans)

```

```python
#H

n = int(input())
arr = list(map(int, input().split()))

ans = [0]*n

tab = 0
car = 0

for i in range(n):
    ans[i] += car
    tab += arr[i]
    car += tab
    
tab = 0
car = 0
mi = ans[-1]

for i in range(n-1, -1, -1):
    mi = min(mi, ans[i]+car)
    tab += arr[i]
    car += tab
    
    mi = min(mi, ans[i]+car)
    
print(mi)
```

```python
#stress check 

import random
from tqdm import tqdm

for _ in tqdm(range(100000)):
    n = random.randint(1, 100)
    arr = [random.randrange(1, 10, 1) for i in range(n)]
    k = random.randint(1, 10)
    irr = [i+1 for i in range(n)]

    
    #solution O(n^2)

    ans1 = []
    for i in irr:
        count = 0
        pnt = i-1
        while(pnt!=0) and (arr[pnt-1] <= arr[pnt]) and (count <= k):
            if arr[pnt-1] == arr[pnt]:
                count += 1
                if count>k:
                    break
            pnt -= 1

        pnt += 1
        ans1.append(pnt)
        
        
    #solution O(n+m)
        
    ans2 = []

    xrr = [0]*n
    count = 0

    for i in range(1, n):
        if arr[i] == arr[i-1]:
            count += 1
            xrr[i] = count

        elif arr[i] < arr[i-1]:
            count = 0
            xrr[i] = -1

    l = 0
    r = 1
    arr[0] = 0

    for r in range(1,n):
        if xrr[r] == -1:
            l = r
            arr[r] = r
            xrr[r] = 0

        elif xrr[r] - xrr[l] <= k:
            arr[r] = l

        else:
            while(xrr[r] - xrr[l] > k):
                l += 1
            arr[r] = l


    for x in irr:
        tmp = arr[x-1]+1
        ans2.append(tmp)

    if ans1 != ans2:
        break
        
print("success")
```
